```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Load Data

```python
mpg = pd.read_csv(r'C:\Users\khiru\OneDrive\Desktop\IS362\auto-mpg.data', header=None, delim_whitespace=True)
mpg
```

TASK 3

```python
col = ['mpg',  'cylinders','displacement','horsepower', 'weight','acceleration','model_yr','origin','car_name']
mpg.columns = col
mpg
```

Task 4

```python
mpg.horsepower = mpg.horsepower.replace('?', np.nan)
mpg.horsepower = pd.to_numeric(mpg.horsepower)
mpg.dtypes
```

Task 5

```python
origins = {1: 'USA', 
          2: 'Asia',
          3: 'Europe'}

mpg.origin = mpg.origin.map(origins)
mpg
```

Task 6

```python
a = sns.countplot(x="cylinders", hue='origin' ,data=mpg)
plt.xlabel("Cylinders",size = 10)
plt.ylabel("Count",size = 10)
```

Task 7

```python
b = sns.regplot(x="horsepower", y="weight", data=mpg, color='y')
plt.title('HorsePower and Weight Relation', fontsize=10)
plt.xlabel("Horsepower",size = 10)
plt.ylabel("Weight",size = 10)
```
